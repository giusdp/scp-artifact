```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV files
requests_tapp = pd.read_csv('tapp_requests.csv')
requests_vanilla = pd.read_csv('vanilla_requests.csv')

# Extract the response_time column
tapp_response_time = requests_tapp['response_time']
vanilla_response_time = requests_vanilla['response_time']

# Calculate statistics
tapp_avg = tapp_response_time.mean()
tapp_median = tapp_response_time.median()
tapp_std = tapp_response_time.std()
tapp_max = tapp_response_time.max()

vanilla_avg = vanilla_response_time.mean()
vanilla_median = vanilla_response_time.median()
vanilla_std = vanilla_response_time.std()
vanilla_max = vanilla_response_time.max()

print(f"Tapp - Average: {tapp_avg}, Median: {tapp_median}, Std Dev: {tapp_std}, Max: {tapp_max}")
print(f"Vanilla - Average: {vanilla_avg}, Median: {vanilla_median}, Std Dev: {vanilla_std}, Max: {vanilla_max}")

# Plot the datasets
plt.plot(tapp_response_time, label='Tapp Response Time')
plt.plot(vanilla_response_time, label='Vanilla Response Time')
plt.xlabel('Request Number')
plt.ylabel('Response Time (ms)')
plt.title('Response Time Comparison')
plt.legend()
plt.show()
```
